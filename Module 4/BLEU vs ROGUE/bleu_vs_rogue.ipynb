```python
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
```

```python
nltk.download("punkt")
```

```python
reference_text = "The cat sat on the mat"
generated_text = "The cat is sitting on the mat"
```

```python
reference_tokens = [reference_text.split()]  # list of references
generated_tokens = generated_text.split()
```

```python
smoothing = SmoothingFunction().method1

bleu_score = sentence_bleu(
    reference_tokens,
    generated_tokens,
    smoothing_function=smoothing
)

print("BLEU Score:", bleu_score)
```

```python
scorer = rouge_scorer.RougeScorer(
    ["rouge1", "rouge2", "rougeL"],
    use_stemmer=True
)
```

```python
rouge_scores = scorer.score(reference_text, generated_text)

for metric, score in rouge_scores.items():
    print(f"{metric}:")
    print("  Precision:", score.precision)
    print("  Recall   :", score.recall)
    print("  F1 Score :", score.fmeasure)
```

```python
references = [
    "AI is transforming healthcare",
    "Transformers changed NLP"
]

generated_outputs = [
    "Artificial intelligence is transforming healthcare",
    "Transformers have revolutionized NLP"
]
```

```python
bleu_scores = []
rouge1_f1 = []
rougeL_f1 = []

for ref, gen in zip(references, generated_outputs):
    # BLEU
    bleu = sentence_bleu(
        [ref.split()],
        gen.split(),
        smoothing_function=smoothing
    )
    bleu_scores.append(bleu)

    # ROUGE
    scores = scorer.score(ref, gen)
    rouge1_f1.append(scores["rouge1"].fmeasure)
    rougeL_f1.append(scores["rougeL"].fmeasure)

print("Average BLEU:", sum(bleu_scores) / len(bleu_scores))
print("Average ROUGE-1 F1:", sum(rouge1_f1) / len(rouge1_f1))
print("Average ROUGE-L F1:", sum(rougeL_f1) / len(rougeL_f1))
```
